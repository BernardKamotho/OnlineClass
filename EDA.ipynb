```python
# create a count plot graph
sns.countplot(x="gender", data=data, palette='spring', hue='city')
plt.title("Distribution of males to females in cities")
plt.xlabel("Gender Distribution")
plt.ylabel("Total number of persons")
```

```python
# Below is a line graph
import matplotlib.dates as mdates

# format the axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b')) # jan feb etc
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

# create the line graph
sns.lineplot(x='date', y='total', data=data)
```

```python
#coming up with pie charts  
#import the libraries 
import pandas as pd
import matplotlib.pyplot as plt
#load your data set
data = pd.read_csv("datasets/supermarket_data.csv")
# Step 1: Count the number of occurrences for each gender
gender_counts = data['payment'].value_counts()
# Step 2: Create the pie chart
plt.pie(
    gender_counts,                  # Values
    labels=gender_counts.index,     # Labels (Male, Female)
    autopct='%1.1f%%',              # Show percentages
    startangle=140,                 # Rotate for better layout
    shadow=True,                    # Add shadow for 3D effect
    colors = ['#00B8A9', '#F6416C', '#FFDE7D', '#FF9A76', '#6A0572', '#AB83A1']
  # Optional: Custom colors
)
# Step 3: Make it circular and add title
plt.axis('equal')
plt.title('Payment Method')
plt.show()
```

```python
# Below is a swarmplot
sns.swarmplot(x='gender', y="total", data=data, palette='winter', hue="payment")
```

```python
# below is a scatter plot
sns.scatterplot(x='rating', y="total", data=data, palette="spring", hue="gender")
```
